```python
import pandas as pd 
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# Read the CSV file
data = pd.read_excel("Example_data.xlsx")

# View the first 5 rows
data.head()
#
#
```

```python
# เปลี่ยนค่าในคอลัมน์ 'Status'
data['Activity'] = data['Activity'].replace({'Active': 1, 'Inactive': 0})
data
```

### count data

```python
# Create a distribution plot using Seaborn
sns.set(style="whitegrid")
plt.figure(figsize=(8, 6))

# Plot the distribution of Antibacterial and non-Antibacterial data
ax = sns.countplot(x='Activity', data=data, palette='viridis')

# Add labels and title
plt.xlabel('Antibacterial Activity')
plt.ylabel('Count')
plt.title('Number of Antibacterial vs Non-Antibacterial Data')

# Annotate the bars with counts
for p in ax.patches:
    ax.annotate(f'{p.get_height()}', (p.get_x() + p.get_width() / 2, p.get_height()),
                ha='center', va='center', xytext=(0, 6), textcoords='offset points')

# Show the plot
plt.show()
```

### Distribution plot

```python
# Filter data for rows where 'Antibacterial' is active
active_antibacterial_data = data[data['Activity'] == 1]

# Extract relevant columns for box plots
selected_data = active_antibacterial_data[['mw', 'polararea', 'xlogp', 'heavycnt', 'hbonddonor', 'hbondacc', 'rotbonds']]

# Create a figure and axis
fig, ax = plt.subplots(figsize=(15, 8))

# Create box plots for each dataset
ax.boxplot(selected_data.values, vert=True, labels=selected_data.columns)

# Set plot labels and title
ax.set_xlabel('Property')
ax.set_ylabel('Value')
ax.set_title('Distribution of Properties for Antibacterial Activity (Active)')

# Show the plot
plt.show()
```

```python
# Filter data for rows where 'Antibacterial' is active
inactive_antibacterial_data = data[data['Activity'] == 0]

# Extract relevant columns for box plots
selected_inact_data = inactive_antibacterial_data[['mw', 'polararea', 'xlogp', 'heavycnt', 'hbonddonor', 'hbondacc', 'rotbonds']]

# Create a figure and axis
fig, ax = plt.subplots(figsize=(15, 8))

# Create box plots for each dataset
ax.boxplot(selected_inact_data.values, vert=True, labels=selected_inact_data.columns)

# Set plot labels and title
ax.set_xlabel('Property')
ax.set_ylabel('Value')
ax.set_title('Distribution of Properties for Antibacterial Activity (Inactive)')

# Show the plot
plt.show()
```

```python
import matplotlib.pyplot as plt


feature_columns = ['mw', 'polararea', 'xlogp', 'heavycnt', 'hbonddonor', 'hbondacc', 'rotbonds']  

label_column = 'Activity'

# Separate data for 'active' and 'inactive' samples
active_data = data[data[label_column] == 1]
inactive_data = data[data[label_column] == 0]

# Loop through each feature to create bar graphs
for feature in feature_columns:
    active_counts = active_data[feature].value_counts()
    inactive_counts = inactive_data[feature].value_counts()

    # Ensure that both 'active' and 'inactive' have the same categories in case some are missing in one group
    all_categories = list(set(active_counts.index) | set(inactive_counts.index))

    # Fill missing categories with 0 count
    for category in all_categories:
        if category not in active_counts:
            active_counts[category] = 0
        if category not in inactive_counts:
            inactive_counts[category] = 0

    # Sort the categories for better visualization
    active_counts = active_counts.sort_index()
    inactive_counts = inactive_counts.sort_index()

    # Plotting the bar graph
    plt.figure(figsize=(10, 8))
    plt.bar(active_counts.index, active_counts.values, width=0.4, align='center', label='Active')
    plt.bar(inactive_counts.index, inactive_counts.values, width=0.4, align='edge', label='Inactive')
    plt.xlabel(feature)
    plt.ylabel('Count')
    plt.title(f'Counts of {feature} for Active and Inactive')
    plt.legend()
    plt.xticks(rotation=45)  # Rotate x-axis labels for better readability
    plt.tight_layout()
    plt.show()
```
