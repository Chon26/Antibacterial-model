```python
import pandas as pd
from rdkit import Chem
from rdkit.Chem import Descriptors

def calculate_descriptors(molecule):
    mol = Chem.MolFromSmiles(molecule)
    if mol is not None:
        mw = Descriptors.MolWt(mol)
        logp = Descriptors.MolLogP(mol)
        num_atoms = Descriptors.HeavyAtomCount(mol)
        num_rotatable_bonds = Descriptors.NumRotatableBonds(mol)
        num_h_donors = Descriptors.NumHDonors(mol)
        num_h_acceptors = Descriptors.NumHAcceptors(mol)
        tpsa = Descriptors.TPSA(mol)


        return {
            "MolecularWeight": mw,
            "LogP": logp,
            "NumAtoms": num_atoms,
            "NumRotatableBonds": num_rotatable_bonds,
            "NumHDonors": num_h_donors,
            "NumHAcceptors": num_h_acceptors,
            "TPSA": tpsa,
            
        }
    else:
        print(f"Error: Unable to generate RDKit molecule from SMILES: {molecule}")
        return None

def calculate_descriptors_for_dataframe(df, smiles_column):
    data = []
    for index, row in df.iterrows():
        compound = row[smiles_column]
        descriptors = calculate_descriptors(compound)
        if descriptors is not None:
            descriptors["Compound"] = compound
            data.append(descriptors)
    return data

# Example usage with a DataFrame
data = {
    "SMILES": ["CCO", "CC(=O)O", "CCN(CC)CC"],
}

df_input = pd.DataFrame(data)

# Specify the column containing SMILES notations
smiles_column_name = "SMILES"

# Calculate descriptors for the DataFrame
results = calculate_descriptors_for_dataframe(df_input, smiles_column_name)

# Convert the results to a DataFrame
df_output = pd.DataFrame(results)

# Display the output DataFrame
df_output
```

```python
import pandas as pd
from rdkit import Chem
from rdkit.Chem import Descriptors

def calculate_all_descriptors(molecule):
    mol = Chem.MolFromSmiles(molecule)
    if mol is not None:
        descriptor_values = {descriptor_name: descriptor_function(mol) for descriptor_name, descriptor_function in Descriptors.descList}
        descriptor_values["SMILES"] = molecule
        return descriptor_values
    else:
        print(f"Error: Unable to generate RDKit molecule from SMILES: {molecule}")
        return None

def calculate_all_descriptors_for_dataframe(df, smiles_column):
    data = []
    for index, row in df.iterrows():
        compound = row[smiles_column]
        descriptors = calculate_all_descriptors(compound)
        if descriptors is not None:
            data.append(descriptors)
    return data

# Example usage with a DataFrame
data = {
    "SMILES": ["CCO", "CC(=O)O", "CCN(CC)CC"],
}

df_input = pd.DataFrame(data)

# Specify the column containing SMILES notations
smiles_column_name = "SMILES"

# Calculate all descriptors for the DataFrame
results = calculate_all_descriptors_for_dataframe(df_input, smiles_column_name)

# Convert the results to a DataFrame
df_output = pd.DataFrame(results)

# Display the output DataFrame
df_output
```
